```python
import pandas as pd
```

```python
df = pd.read_json('./scrapyProjects/skillsNaukri/skillsNaukri/spiders/skillsNaukri.json')
```

```python
df.head()
```

```python
skillList = []
```

```python
for i in range(len(df)):
    skillList.extend(df.loc[i])
```

```python
skillList
```

```python
skillListNew=[]
```

```python
for lis in skillList:
    skillListNew.extend(lis)
```

```python
skillListNew
```

```python
skillListNew2 = []
```

```python
for skillLis in skillListNew:
    skillListNew2.extend(skillLis.split(','))
```

```python
skillListNew2
```

```python
skillSeries = pd.Series(data=skillListNew2)
```

```python
df2 = pd.DataFrame()
```

```python
df2['Skill'] = skillSeries
```

```python
df3 = df2['Skill'].str.replace('\.\.\.','')
```

```python
df3.to_csv('./scrapyProjects/skillsNaukri/skillsNaukri/spiders/skillsNaukri.csv',index=False)
```
